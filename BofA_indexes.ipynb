```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import datetime
from tqdm.notebook import tqdm
```

```python
# ежедневная загрузка индексов
def df_for_concat(lst,dates,id):
    data_for_concat = {'Date':dates,
                      'ID':id,
                      'Value':lst}
    df = pd.DataFrame(data_for_concat)
    return df


PROXY = "ip-адрес:порт"

chrome_options = Options()
chrome_options.add_argument(f"--proxy-server=http://{PROXY}")
driver = webdriver.Chrome(options=chrome_options)
driver.get('https://indices.ice.com/')

time.sleep(5)
# принимаем правила
accept_button = driver.find_element(By.CSS_SELECTOR, '#btnAccept')
accept_button.click()

time.sleep(10)

# открываем поле с валютами
field_to_click = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.indexSnapshotField > div:nth-child(3) > div > div > div > div")
field_to_click.click()

time.sleep(10) #меняем валюту вручную

total_returns_list = []
ytds_list = []
eff_yields_list = []
spreads_list = []
eff_durs_list = []

# меняем тикер
df_tickers = pd.read_excel('tickers.xlsx')
tickers = df_tickers['Tickers'].to_list()
for t in tickers:
    
    ticker = t
    input_field = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.indexSnapshotField > div:nth-child(1) > div > div > input")
    input_field.clear()
    input_field.send_keys(ticker)
    input_field.send_keys(Keys.RETURN)
    
    time.sleep(10)
    

    dates = ['10/03/2024','10/04/2024']
    
    total_returns = []
    ytds = []
    eff_yields = []
    spreads = []
    eff_durs = []

    for d in dates:
        # меняем дату
        date = d
        date_field = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.indexSnapshotField > div:nth-child(2) > div > div > input")
        date_field.clear()
        date_field.send_keys(date)
        date_field.send_keys(Keys.RETURN)
        
        time.sleep(12)
        # total return
        value_field = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.snapshotGrid.css-odwlz2 > div > div > div > div.ag-root-wrapper-body.ag-layout-normal.ag-focus-managed > div.ag-root.ag-unselectable.ag-layout-normal > div.ag-body-viewport.ag-layout-normal.ag-row-no-animation > div.ag-center-cols-clipper > div > div > div.ag-row.ag-row-no-focus.ag-row-even.ag-row-level-0.ag-row-position-absolute.ag-row-first > div:nth-child(6)")
        total_return = float(value_field.text)
        total_returns.append(total_return)
        
        # ytd
        value_field_2 = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.snapshotGrid.css-odwlz2 > div > div > div > div.ag-root-wrapper-body.ag-layout-normal.ag-focus-managed > div.ag-root.ag-unselectable.ag-layout-normal > div.ag-body-viewport.ag-layout-normal.ag-row-no-animation > div.ag-center-cols-clipper > div > div > div.ag-row.ag-row-no-focus.ag-row-even.ag-row-level-0.ag-row-position-absolute.ag-row-first > div:nth-child(8)")
        ytd = float(value_field_2.text)
        ytds.append(ytd)
        
        # Eff Yield
        value_field_3 = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.snapshotGrid.css-odwlz2 > div > div > div > div.ag-root-wrapper-body.ag-layout-normal.ag-focus-managed > div.ag-root.ag-unselectable.ag-layout-normal > div.ag-body-viewport.ag-layout-normal.ag-row-no-animation > div.ag-center-cols-clipper > div > div > div:nth-child(3) > div:nth-child(2)")
        eff_yield = float(value_field_3.text)
        eff_yields.append(eff_yield)
        
        # Spread
        value_field_4 = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.snapshotGrid.css-odwlz2 > div > div > div > div.ag-root-wrapper-body.ag-layout-normal.ag-focus-managed > div.ag-root.ag-unselectable.ag-layout-normal > div.ag-body-viewport.ag-layout-normal.ag-row-no-animation > div.ag-center-cols-clipper > div > div > div:nth-child(3) > div:nth-child(4)")
        spread = float(value_field_4.text)
        spreads.append(spread)
    
        # Eff Dur
        value_field_5 = driver.find_element(By.CSS_SELECTOR, "#container > div > div > div > div.css-198fgti > div.css-dyiqfq > div > div > div.css-1g1p646 > div > div.css-90yehr > div.css-qrvl73 > div > div.leftPane > div.homeSnapshot.css-1jh5xym > div.overlayContainer > div.snapshotGrid.css-odwlz2 > div > div > div > div.ag-root-wrapper-body.ag-layout-normal.ag-focus-managed > div.ag-root.ag-unselectable.ag-layout-normal > div.ag-body-viewport.ag-layout-normal.ag-row-no-animation > div.ag-center-cols-clipper > div > div > div:nth-child(4) > div:nth-child(2)")
        eff_dur = float(value_field_5.text)
        eff_durs.append(eff_dur)
        
        time.sleep(5)

    total_returns_list.append(total_returns)
    ytds_list.append(ytds)
    eff_yields_list.append(eff_yields)
    spreads_list.append(spreads)
    eff_durs_list.append(eff_durs)


# формируем датафреймы с данными
#total return 
ttl_rtrns_ids = df_tickers['Total_return_id'].to_list()
dfs_to_concat_tr = []
for i, ticker in enumerate(tickers):
    df_total_return = df_for_concat(total_returns_list[i], dates, ttl_rtrns_ids[i])
    dfs_to_concat_tr.append(df_total_return)
    
df_concatenated_tr = pd.concat(dfs_to_concat_tr, ignore_index=True)

# ytd
ytd_ids = df_tickers['YTD_id'].to_list()
dfs_to_concat_ytd = []
for i, ticker in enumerate(tickers):
    df_ytd = df_for_concat(ytds_list[i], dates, ytd_ids[i])
    dfs_to_concat_ytd.append(df_ytd)
    
df_concatenated_ytd = pd.concat(dfs_to_concat_ytd, ignore_index=True)

#eff yield
eff_yield_ids = df_tickers['Eff_Yield_id'].to_list()
dfs_to_concat_eff_yield = []
for i, ticker in enumerate(tickers):
    df_eff_yield = df_for_concat(eff_yields_list[i], dates, eff_yield_ids[i])
    dfs_to_concat_eff_yield.append(df_eff_yield)
    
df_concatenated_eff_yield = pd.concat(dfs_to_concat_eff_yield, ignore_index=True)

# spread
spread_ids = df_tickers['Spread-Govt_id'].to_list()
dfs_to_concat_spread = []
for i, ticker in enumerate(tickers):
    df_spread = df_for_concat(spreads_list[i], dates, spread_ids[i])
    dfs_to_concat_spread.append(df_spread)
    
df_concatenated_spread = pd.concat(dfs_to_concat_spread, ignore_index=True)

# eff duration
eff_dur_ids = df_tickers['Eff_duration_id'].to_list()
dfs_to_concat_eff_dur = []
for i, ticker in enumerate(tickers):
    df_eff_dur = df_for_concat(eff_durs_list[i], dates, eff_dur_ids[i])
    dfs_to_concat_eff_dur.append(df_eff_dur)
    
df_concatenated_eff_dur = pd.concat(dfs_to_concat_eff_dur, ignore_index=True)

# объединяем датафреймы в один
df_concatenated = pd.concat([df_concatenated_tr,
                             df_concatenated_ytd,
                             df_concatenated_eff_yield,
                             df_concatenated_spread,
                            df_concatenated_eff_dur],ignore_index=True)

df_concatenated['Date'] = pd.to_datetime(df_concatenated['Date'], format='%m/%d/%Y').dt.strftime('%d.%m.%Y')
df_concatenated.to_excel('BofA_data.xlsx',header=False,index=False)
print(df_concatenated)
```
